```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x=np.arange(10)
y=x**2
plt.plot(x,y)
plt.show()
```

```python
x=np.arange(10)
y=x**2
plt.plot(x,y,'b+')
plt.show()
```

```python
x=np.arange(10)
y=x**2
plt.plot(x,y,'b-')
plt.show()
```

```python
x=np.arange(10)
y=x**2
plt.plot(x,y,'bo')
plt.show()
```

```python
x=np.arange(10)
y=x**2
plt.plot(x,y,'ko')
plt.show()
```

```python
x=np.arange(10)
y=x**2
plt.plot(x,y,'ko',x,y,'b-')
plt.show()
```

```python
x=np.arange(10)
y=x**2
plt.plot(x,y,'bo',x,y,'k-')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Scatter and Line of x2')


plt.xlim([0,10])
plt.ylim([0,100])
plt.show()
```

```python
plt.plot([1,3,4,6,9],[1,8,3,5,1])
plt.xlabel('x')
plt.ylabel('y')
plt.title('Scatter and Line')


plt.xlim([0,10])
plt.ylim([0,10])
plt.show()
```

```python
x=np.arange(1,21)
y=np.log(x)
z=np.exp(x)
#figure initialization
plt.figure()

plt.subplot(1,2,1)
plt.title("x=log x")
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('log(x)')


plt.subplot(1,2,2)
plt.title('x=exp(x)')
plt.plot(x,z)
plt.xlabel('x')
plt.ylabel('exp(x)')
plt.show()
```
